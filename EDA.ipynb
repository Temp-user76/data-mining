```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("C:\\Users\\nitin\\Downloads\\Fremont_Bridge_Bicycle_Counter_20250414.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df['total']=df[['Fremont Bridge Sidewalks, south of N 34th St Total','Fremont Bridge Sidewalks, south of N 34th St Cyclist West Sidewalk','Fremont Bridge Sidewalks, south of N 34th St Cyclist East Sidewalk']].sum(axis=1)
```

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df['hour']=df['Date'].dt.hour
df['dayofweek']=df['Date'].dt.dayofweek
df['year']=df['Date'].dt.year
df['month']=df['Date'].dt.month
```

```python
df.head()
```

```python
df=df.drop(['Date'],axis=1)
```

```python
df.columns
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df['total'].plot(figsize=(15,5),title='Daily Bike counts')
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='hour', y='total', data=df)
plt.title("Hourly Bike count pattens")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='dayofweek', y='total', data=df)
plt.title("Weekday Commute Trends")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,r2_score,f1_score
```

```python
x = df[['hour', 'dayofweek', 'month']]
y = df['total']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=2)
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
print("R Score:", r2_score(y_test, y_pred))
```

```python
sample = [[8, 2, 4]]  # Example: 8 AM, Wednesday, April
prediction = lr.predict(sample)
print(f"Predicted bike count: {int(prediction[0])}")
```
